```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_ta as ta
```

```python
df = pd.read_csv("BTC-USDT-1m-clean.csv", index_col=0, parse_dates=[0])
```

```python
df.tail()
```

```python
df["ema"] = df.ta.ema(length=200)
```

```python
df["ohlc4"] = (df["open"] + df["high"] + df["low"] + df["close"]) / 4
```

```python
idx = np.random.randint(0, len(df))
```

```python
c = df.iloc[idx-29:idx + 1].ohlc4
h = df.iloc[idx-29:idx + 1].high
```

```python
np.log(c) - np.log(c[-1])
```

```python
# make 3 row 1 col subplots
fig, ax = plt.subplots(3, 1, figsize=(10, 10))

ax[0].plot(c.index, c, label="close")
ax[0].plot(h.index, h, label="high")

ax[1].plot(c.index, np.log(c) - np.log(c[-1]), label="diff from last")
```
